## Marker detection

```python
marker_dict=aruco.getPredefinedDictionary(aruco.DICT_4X4_50)
para_markers=aruco.DetectorParameters()

cap=cv2.VideoCapture(0)


while True:
    ret,frame=cap.read()
    if ret==False:
        break
    gray_img=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    marker_corners,marker_ids,reject=aruco.detectMarkers(gray_img,marker_dict,parameters=para_markers)
    if marker_corners:
        x=marker_corners[0][0][1]
        print(x)
#         print(len(marker_ids))
#         print(marker_ids)
        for j in range(len(marker_ids)):
            for i in range(4):
                cv2.line(frame,(int(marker_corners[j][0][i-1][0]),int(marker_corners[j][0][i-1][1])),(int(marker_corners[j][0][i][0]),int(marker_corners[j][0][i][1])),(0,255,0),4)
    cv2.imshow("frame",frame)
    
    key=cv2.waitKey(1)
    if key==ord("q"):
        break
cap.release()
cv2.destroyAllWindows()
        
        
```

# Camera Calibration

```python
import os
```

```python
import time
```

```python
image_dir_path = "calibration_images"

check_dir = os.path.isdir(image_dir_path)
```

```python
if not check_dir:
    os.makedirs(image_dir_path)
```

```python
board_dim=(9,6)
```

```python
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 30, 0.001)
def detect_checker_board(image, grayImage, criteria, boardDimension):
    ret, corners = cv2.findChessboardCorners(grayImage, boardDimension)
    if ret == True:
        corners1 = cv2.cornerSubPix(grayImage, corners, (3, 3), (-1, -1), criteria)
        image = cv2.drawChessboardCorners(image, boardDimension, corners1, ret)

    return image, ret
```

```python
capt=cv2.VideoCapture(1,cv2.CAP_DSHOW)
capt.set(cv2.CAP_PROP_FRAME_WIDTH, 10000)
capt.set(cv2.CAP_PROP_FRAME_HEIGHT, 10000)
capt.set(cv2.CAP_PROP_FPS, 60)
n=0
time_prev = time.time()
while True:
    ret,frame=capt.read()
#     copy_frame=frame.copy()
    if ret==False:
        break
    gray_img=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    
    image,board_detected=detect_checker_board(frame, gray_img, criteria, board_dim)
    
    cv2.putText(image, f"image_{n}", (50,40), cv2.FONT_HERSHEY_PLAIN, 2, (0,0,255))
    cv2.imshow("frame", image)
    key=cv2.waitKey(1)
    
    if board_detected:
        if(n>30):
            break
        if(time.time() - time_prev >= 5):
            cv2.imwrite(f"{image_dir_path}/image_{n}.png",frame)
            n+=1
            time_prev = time.time()                
    
    if key==ord("s"):
        n+=1
        cv2.imwrite(f"{image_dir_path}/image_{n}.png",frame)
    if key==ord("q"):
        print(f"{n} images are saved")
        break
    
capt.release()
cv2.destroyAllWindows()
    
```

```python
import os
import cv2
from cv2 import aruco
import numpy as np
```

```python
chessboard_dim=(9,6)
square_size=2.4
Criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 30, 0.001)

calibdata_path="camera_calibdata"
checkdir =os.path.isdir(calibdata_path)

# os.makedirs(calibdata_path)
# ----------------------------------------------
obj_3D = np.zeros((chessboard_dim[0] * chessboard_dim[1], 3), np.float32)
print(obj_3D)
obj_3D[:, :2] = np.mgrid[0 : chessboard_dim[0], 0 : chessboard_dim[1]].T.reshape(
    -1, 2
)
print("**************************************************")
print(obj_3D)
obj_3D *= square_size
print("**************************************************")
print(obj_3D)

obj_points_3d=[]
img_points_2d=[]
count=0
image_dir_path="calibration_images"
files=os.listdir(image_dir_path)
for file in files[1:]:
    imagepath = os.path.join(image_dir_path, file)
    img=cv2.imread(imagepath)
    print(imagepath)
    gray_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    ret,corners=cv2.findChessboardCorners(img, chessboard_dim, None, cv2.CALIB_CB_ADAPTIVE_THRESH + cv2.CALIB_CB_NORMALIZE_IMAGE
        + cv2.CALIB_CB_FAST_CHECK)
    if ret==True:
        obj_points_3d.append(obj_3D)
        corners2 = cv2.cornerSubPix(gray_img, corners, (3, 3), (-1, -1), Criteria)
        img_points_2d.append(corners2)
        image = cv2.drawChessboardCorners(img, chessboard_dim, corners2, ret)
        count+=1
cv2.destroyAllWindows()
print(obj_points_3d)
print("********************************************************")
print(count)
print("********************************************************")
print(img_points_2d)
```

```python
class Queue:
    class Node:
        __slots__ = '_element','_next'
        def __init__(self,data,next):
            self._element=data
            self._next=next

    def __init__(self):
        self._head=None
        self._tail=None
        self._size=0

    def _len(self):
        return self._size

    def isEmpty(self):
        return self._size==0

    def _print(self):
        t=self._head
        for i in range(self._size):
            if t is not None:
                print(t._element)
                t=t._next

    def enqueue(self,data):
        n=self.Node(data,None)
        if self.isEmpty():
            self._head=n
            self._tail=n
        else:
            (self._tail)._next=n
            self._tail=n

        self._size+=1

    def dequeue(self):
        if self._size!=0:
            t=self._head
            self._head=t._next
            self._size-=1
            # return t._element
        else:
            return
```
